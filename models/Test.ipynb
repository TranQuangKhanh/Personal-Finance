```python
import numpy as np
import joblib

# Load all saved artifacts
gmm = joblib.load('../models/gmm_model.pkl')
gmm_scaler = joblib.load('../models/gmm_scaler.pkl')
label_map = joblib.load('../models/gmm_label_map.pkl')

rf = joblib.load('../models/rf_model.pkl')
rf_scaler = joblib.load('../models/rf_scaler.pkl')
le = joblib.load('../models/rf_label_encoder.pkl')

print('All models loaded.')
```

```python
def predict_user(user_input: dict) -> dict:
    """
    Input: dict with financial features
    Output: dict with segment + cash_flow_status
    """
    # GMM prediction
    gmm_features = ['monthly_income', 'savings_rate', 'debt_to_income_ratio']
    gmm_input = np.array([[user_input[f] for f in gmm_features]])
    gmm_input_scaled = gmm_scaler.transform(gmm_input)
    cluster = gmm.predict(gmm_input_scaled)[0]
    segment = label_map[cluster]

    # RF prediction
    rf_features = [
        'monthly_income', 'monthly_expense_total', 'savings_rate',
        'debt_to_income_ratio', 'credit_score', 'loan_payment',
        'investment_amount', 'discretionary_spending', 'essential_spending',
        'emergency_fund'
    ]
    rf_input = np.array([[user_input[f] for f in rf_features]])
    rf_input_scaled = rf_scaler.transform(rf_input)
    cash_flow = le.inverse_transform(rf.predict(rf_input_scaled))[0]

    return {
        'segment': segment,
        'cash_flow_status': cash_flow
    }

print('predict_user() ready.')
```

```python
test_1 = {
    'monthly_income': 2500,
    'monthly_expense_total': 2800,
    'savings_rate': 0.05,
    'debt_to_income_ratio': 0.65,
    'credit_score': 580,
    'loan_payment': 400,
    'investment_amount': 0,
    'discretionary_spending': 600,
    'essential_spending': 1800,
    'emergency_fund': 200
}

result_1 = predict_user(test_1)
print('Test 1 Result:')
print(f"  Segment       : {result_1['segment']}")
print(f"  Cash Flow     : {result_1['cash_flow_status']}")
```
